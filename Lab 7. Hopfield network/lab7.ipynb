```python
b_formated = [x.replace(' ', '').strip('\n') for x in b]
test_array = [[(1 if y == '1' else -1) for y in x.replace('\n', '')] for x in b_formated]
```

```python
x_arr = num_array
# емкость (количество запоминаемых образов)
L = len(x_arr)
K = len(x_arr[0])
```

```python
def func(net_arr, old_net_arr):
    new_net = []
    for x in net_arr:
        if x > 0:
            new_net.append(1)
        elif x < 0:
            new_net.append(-1)
        else:
            new_net.append(func(old_net_arr[-1], old_net_arr[:-1])[x])
    return new_net
```

```python
def epoxa(w_arr, y_arr, net_arr, prev_net_arr):
    for k in range(K):
        net_arr[k] = sum([w_arr[j][k] * y_arr[j] for j in range(K)])
    y_arr = func(net_arr, prev_net_arr)
    return net_arr, y_arr
```
